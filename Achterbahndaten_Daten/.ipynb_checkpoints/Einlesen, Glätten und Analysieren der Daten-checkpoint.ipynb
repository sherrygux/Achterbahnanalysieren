```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.ndimage import filters
from scipy.optimize import minimize
import json
```

```python
df1 = pd.read_csv('./loops/coastersmall_formatted.csv', usecols=[0,1,2,3])
#df2 = pd.read_csv('./loops/loop1_formatted.csv', usecols=[0,1,2,3])
#df3 = pd.read_csv('./loops/loop2_formatted.csv', usecols=[0,1,2,3])
data1 = np.array(df1)
#data2 = np.array(df2)
#data3 = np.array(df3)
```

```python
t_N = data1[-1][0]
N = data1.shape[0]
h = 1/N
```

```python
t_interp = np.linspace(0, (N-1)*h*t_N, N, endpoint=True)
xp = data1[:,0]
fp_x = data1[:,1]
fp_z = data1[:,3]
x_y_axis = np.interp(t_interp, xp, fp_x)
z_y_axis = np.interp(t_interp, xp, fp_z)
x_y_axis_filter = filters.gaussian_filter(x_y_axis, 5)
z_y_axis_filter = filters.gaussian_filter(z_y_axis, 5)
```

```python
plt.axis([12,26,-1.0,1.5])
plt.plot(t_interp, x_y_axis_filter, color='r')
plt.plot(t_interp, z_y_axis_filter, color='b')
plt.show()
```

```python
def find_ellipsen(t_N0, t_N0pN):
    
    def find_interval(t_N0, t_N0pN):
        return np.array([t for t in t_interp if t >= t_N0 and t <= t_N0pN])
    
    def loss(args):
        E = 0
        x_m, z_m, r, s = args
        for t in interval:
            a_ref = np.array(
                [x_m + r * np.sin(2 * np.pi * (t - t_N0)/(t_N0pN - t_N0)), 
                 z_m + s * np.cos(2 * np.pi * (t - t_N0)/(t_N0pN - t_N0))])
            a = np.array(
                [x_y_axis_filter[list(t_interp).index(t)], 
                 z_y_axis_filter[list(t_interp).index(t)]])
            E += np.linalg.norm(a - a_ref) ** 2
        return E / (N + 1)
    
    interval = find_interval(t_N0, t_N0pN)
    N = interval.shape[0] - 1
    x0 = np.random.rand(4) 
    bnds = ((None,None),(None,None),(0,None),(0,None))
    res = minimize(loss, x0, bounds=bnds)
    x_m, z_m, r, s = res.x
    if res.fun/N<=0.005 and r>0.5 and s>0.8:
        ellipsen_x = np.zeros((N + 1,))
        ellipsen_z = np.zeros((N + 1,))
        for i, t in enumerate(interval):
            ellipsen_x[i] = x_m + r * np.sin(2 * np.pi * (t - t_N0) / (t_N0pN - t_N0))
            ellipsen_z[i] = z_m + s * np.cos(2 * np.pi * (t - t_N0) / (t_N0pN - t_N0))
        return res.fun/N, interval, ellipsen_x, ellipsen_z 
```

```python
t_splited = np.array_split(t_interp, 6)
for search_intervall in t_splited:
    loss_min = 999
    interval_opt = None
    ellipsen_x_opt = None
    ellipsen_z_opt = None
    isplot = False
    for i in range(1,len(search_intervall)):
        if find_ellipsen(search_intervall[0], search_intervall[i]):
            isplot = True
            loss, interval, ellipsen_x, ellipsen_z =  find_ellipsen(search_intervall[0], search_intervall[i])
            if  loss < loss_min:
                loss_min = loss
                interval_opt = interval
                ellipsen_x_opt = ellipsen_x
                ellipsen_z_opt = ellipsen_z
    if isplot:
        plt.axis([12,26,-1.0,1.5])
        plt.plot(t_interp, x_y_axis_filter, color='r')
        plt.plot(t_interp, z_y_axis_filter, color='b')
        plt.plot(interval_opt, ellipsen_x_opt, color='g')
        plt.plot(interval_opt, ellipsen_z_opt, color='m')
        plt.show()
```

```python
def calibration(accs_file, json_file):
    
    def rotation(n, alpha):
        aa = 1 - np.cos(alpha)
        bb = np.sin(alpha)
        cc = np.cos(alpha)
        return np.array([[n[0] ** 2 * aa + cc, n[0] * n[1] * aa - n[2] * bb, n[0] * n[2] * aa + n[1] * bb],
                           [n[0] * n[1] * aa + n[2] * bb, n[1] ** 2 * aa + cc, n[1] * n[2] * aa - n[0] * bb],
                           [n[0] * n[2] * aa - n[1] * bb, n[1] * n[2] * aa + n[0] * bb, n[2] ** 2 * aa + cc]])

    accs = np.array(pd.read_csv(accs_file, usecols=[1,2,3,4]))
    t_coaster = accs[:,0]
    #plt.figure(figsize=(30,10))
    #plt.axis([0,300,-3,1.5])
    #plt.plot(t_coaster, accs[:,1], color='r')
    #plt.plot(t_coaster, accs[:,2], color='g')
    #plt.plot(t_coaster, accs[:,3], color='b')
    #plt.show()
    
    with open(json_file,"r") as load_f:
        load_dict = json.load(load_f)
        zhat_interval = load_dict['known_orientations']['zhat']
        zx_interval = load_dict['known_orientations']['z-x']
        
    t_begin = np.array([t for t in t_coaster if t>=zhat_interval[0] and t<=zhat_interval[1]])
    z = accs[list(t_coaster).index(t_begin[0]):list(t_coaster).index(t_begin[-1]),1:]
    zhat = np.mean(z, axis=0)
    zhat_norm = zhat/np.linalg.norm(zhat)
    e_3 = np.array([0,0,1])
    a = np.cross(zhat_norm, e_3)
    a_norm = a/np.linalg.norm(a)
    phi = np.arctan2(np.linalg.norm(a), np.inner(zhat_norm, e_3))
    
    accs_new = accs
    accs_new[:,1:] = (rotation(a_norm, phi) @ accs[:,1:].T).T
    
    #plt.plot(t_coaster, accs_new[:,1], color='r')
    #plt.plot(t_coaster, accs_new[:,2], color='g')
    #plt.plot(t_coaster, accs_new[:,3], color='b')
    #plt.show()
    
    t_rise = np.array([t for t in t_coaster if t>=zx_interval[0] and t<=zx_interval[1]])
    v = accs_new[list(t_coaster).index(t_rise[0]):list(t_coaster).index(t_rise[-1]),1:]
    vhat = np.mean(v, axis=0)
    vhat_norm = vhat/np.linalg.norm(vhat)
    psi = -np.arctan2(vhat_norm[1], vhat_norm[0])
    accs_final = accs_new
    accs_final[:,1:] = (rotation(e_3, psi) @ accs_new[:,1:].T).T
   
    accs_abs = np.sqrt(accs_final[:,1] ** 2 + accs_final[:,2] ** 2 + accs_final[:,3] ** 2)
    
    #plt.figure(figsize=(20,10))
    #plt.axis([0,300,-2,4])
    #plt.plot(t_coaster, accs_final[:,1], color='r')
    #plt.plot(t_coaster, accs_final[:,2], color='g')
    #plt.plot(t_coaster, accs_final[:,3], color='b')
    #plt.plot(t_coaster, accs_abs, color='k')
    #plt.show()
    
    return accs_final, accs_abs, t_coaster
```

```python
def max_accs(accs_file, json_file):
    data = calibration(accs_file, json_file)[0]
    maximum = np.max(data[:,1:],axis=0)
    maximum_index = np.argmax(data[:,1:],axis=0)
    t_x = data[maximum_index[0]][0]
    t_y = data[maximum_index[1]][0]
    t_z = data[maximum_index[2]][0]
    return maximum, t_x, t_y, t_z
```

```python
max_accs("./coasters/magnum/accelerometer_log.txt", "./coasters/magnum/notes.json")
```

```python
def frequency_accs(accs_file, json_file):
    data = calibration(accs_file, json_file)[0]
    accs_abs = calibration(accs_file, json_file)[1]
    N = data.shape[0]
    freq = 0
    maximum = np.max(accs_abs,axis=0)
    if maximum < 2:
        print("This coaster is not so fast!")
    else:
        for accs in accs_abs:
            if accs >= 0.7 * maximum:
                freq += 1
    return freq/N
```

```python
frequency_accs("./coasters/magnum/accelerometer_log.txt", "./coasters/magnum/notes.json")
```

```python
def weightlessness(accs_file, json_file):
    data = calibration(accs_file, json_file)[0]
    N = data.shape[0]
    i = 0
    for accs_z in data[:,3]:
        if accs_z >= 1:
            i += 1
    return i/N
```

```python
weightlessness("./coasters/magnum/accelerometer_log.txt", "./coasters/magnum/notes.json")
```

```python
def angle_hill(accs_file, json_file):
    with open(json_file,"r") as load_f:
        load_dict = json.load(load_f)
        zx_interval = load_dict['known_orientations']['z-x']
    data = calibration(accs_file, json_file)[0]
    t_coaster = calibration(accs_file, json_file)[2]
    t_rise = np.array([t for t in t_coaster if t>=zx_interval[0] and t<=zx_interval[1]])
    v = data[list(t_coaster).index(t_rise[0]):list(t_coaster).index(t_rise[-1]),1:]
    vhat = np.mean(v, axis=0)
    print(np.arctan2(vhat[0], vhat[2])*180/np.pi)
    
    
```
